```python
#------------------------------------------------------------------------------------------------------------#
#
# Première constatation des outliers:
#
# Les champs numériques suivants montrent des valeurs minimales négatives ce qui est une erreur
#
#  trans-fat_100g
#  sugars_100g
#  fiber_100g
#  proteins_100g
#  vitamin-a_100g
#  vitamin-c_100g
#  iron_100g
#  copper_100g
#  selenium_100g
#
#------------------------------------------------------------------------------------------------------------#
```

```python
#------------------------------------------------------------------------------------------------------------#
#
# Nombre de valeurs négatives par colonne - on garde la liste dans la variable col_below
#
#------------------------------------------------------------------------------------------------------------#

cols = food_data.columns

cols_below = []            # Garde la liste de noms de colonnes avec des valeurs negatives

for col_name in cols:
    if food_data.dtypes[col_name]==float:
        below_zero = food_data[col_name].lt(0).sum()       
        print('Column: ',col_name,' - Nombre de valeurs négatives: ', below_zero)
        if below_zero > 0:
            cols_below.extend([col_name])
```

```python
#------------------------------------------------------------------------------------------------------------#
#
# À partir du rapport précédent, on obtient que les colonnes qui ont des valeurs
# négatives aberrantes sont les suivantes:
#
#  trans-fat_100g : 4 values
#  sugars_100g: 7 values
#  fiber_100g: 1 value
#  proteins_100g: 3 values
#  vitamin-a_100g: 1
#  vitamin-c_100g: 1
#  iron_100g: 1
#  copper_100g: 1
#  selenium_100g: 1
#
# En revanche les deux colonnes suivantes sont censées de contenir 
# parfois des valeurs négatives:
#
#  nutrition-score-fr_100g
#  nutrition-score-uk_100g
#
# On va les analyser séparément pour prendre en suite la décision sur 
# la gestion des valeurs négatives
#
# J'efface les deux colonnes suivantes de la liste pour povoir faire des itérations 
# après sur cette liste
#
#------------------------------------------------------------------------------------------------------------#

cols_below.remove('nutrition-score-fr_100g')

cols_below.remove('nutrition-score-uk_100g')

cols_below
```

```python
#------------------------------------------------------------------------------------------------------------#
#
#  Description des données - les contenus des registres avec des valeurs négatives 
#  dans des colonnes d'information nutritionnelle, afin de les vérifier dans la base 
#  de données officielle de OpenFoodData
#
#------------------------------------------------------------------------------------------------------------#

cols_below
for col in cols_below:
    print('\n',col)
    print(food_data['code'][food_data[col]<0 ],
          '\n',food_data['product_name'][food_data[col]<0 ],
          '\n',food_data[col][food_data[col]<0]
         )
    
```

```python
#------------------------------------------------------------------------------------------------------------#
#
#
#  D'après des vérifications faites sur internet (notament la base de données 
#  officielle de OpenFoodData), les valeurs négatives dans des colonnes d'information
#  nutritionnelle sont PARFOIS remplacées par ZÉRO.
#
#  Dans le site officiel de OpenFoodData les valeurs négatives apparaissent dans 
#  certains cas. 
#
#  Par exemple, le code = 0032971251990 montre une valeur négative dans la colonne 
#  selenium_100g.
#
#  Donc, pour chacune de colonnes ci-dessus les valeurs négatives seront remplacées 
#  par zéro.
#
#  Maintenant, on remplacee toutes les valeurs négatives en zéros dans les colonnes 
#  d'informations nutritionnelles
#
#  La variable cols_below contient déjà les noms des colonnes à changer
#
#------------------------------------------------------------------------------------------------------------#

for col in cols_below:
    food_data[col] = food_data[col].map(lambda x: x if x > 0 else 0)

print('Ready')
```

```python
#---------------------------------------------------------------------------------------------------------------------------------
#
# Nous passons maintenant à analyser les outliers par excès (les valeurs négatives ont
# été déjà remis 'a zéro), des valeurs supérieures extrèmes.
#
# À partir des boîtes à moustaches (concernant les colonnes avec des valeurs négatives),
# on remarque que les colonnes présentent des anomalies des valeurs extrêmes. 
# On fera une première detection des outliers avec les BoxPLot pour chacune des 
# colonnes numériques.
#
# Une méthode de détecter des outliers est la suivante:
#
# la valeur est consideré comme un outlier si elle est supérieure ou inferieure
# à K fois l'écart interquartile. La litterature conseille de prendre K = 1.5
#
# Je ferai un test sur une première colonne afin de developper la méthode correcte 
# pour toutes les autres.
#
#---------------------------------------------------------------------------------------------------------------------------------
```
