```python
#
# les 9 colonnes suivantes seront eliminées du dataframe car elles sont nulles pour tous les registres
#
null_columns = ['ph_100g', 'collagen-meat-protein-ratio_100g', 'chlorophyl_100g', 'carbon-footprint_100g',
               'glycemic-index_100g', 'water-hardness_100g', 'phylloquinone_100g',
                'beta-glucan_100g', 'carnitine_100g']
for null_column in null_columns:
    data.drop(null_column, axis=1, inplace=True)
data.shape
```

```python
#
# lire dans un dataframe les noms possibles de la France en plusieurs langues - fichier france_names_latin_alphabets.csv
#
france_names = pd.read_csv('france_names_latin_alphabets.csv',header=None)
#
# on met tous en minuscule
#
france_names[0] = france_names[0].apply(lambda x: x.lower())
#
# on enkève les blancs au début età la fin
#
france_names[0] = france_names[0].apply(lambda x: x.strip())
france_names
```

```python
#
# Convertir tous les pays de la colonne "countries" du dataframe en minuscule
#

data["countries"] = data["countries"].map(lambda x: x if type(x)!=str else x.lower())
```

```python
list_of_countries = data.groupby(['countries']).count()
print(list_of_countries)
```

```python
#
# On eliimine du dataframe les produits que n'ont pas la France dans leur colonne "countries"
#
# D'abord, on passe tous les noms de France vers une liste pour que ce soit facil de faire des itérations
#
valid_codes_france = france_names[0].tolist()
#
# print(valid_codes_france)
#
# On choisit les lignes du dataframe qui ont la France dans leur liste de "countries"
#
# data 3 stocke tous les produits dont la colonne "countries" contient la France
#
pattern = '|'.join(valid_codes_france)
data = data.loc[(data['countries'].str.contains(pattern, case=False, na=False))]
```
